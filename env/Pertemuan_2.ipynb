```python
#elif
x = 80

if x < 50:
    print("x lebih kecil dari 50")
elif x > 100:
    print("x lebih besar dari 100")
else:
    print("x lebih besar dari 50 dan lebih kecil sama dengan dari 100")
```

```python
#contoh overlapping, karena dibaca dari atas
x = 20

if x < 50:
    print("x lebih kecil dari 50")
elif x < 20:
    print("x lebih kecil dari 20")
else:
    print("x lebih besar dari 50")

print("lalu print")
```

While Loops

```python
while <condition>:
    <jalanin ini>
```

```python
#infinite loop
while True:
    primt("hello")
```

```python
n = 5
while n > 0:
    print("n awal", n)
    n -= 1
    #print("n akhir", n)
```

```python
i = 0
while i < 6:
    print("i awal", i)
    i += 1
    #print("i akhir", i)
```

Break

#### break
keluar dari Lloop

#### continue
lanjut/loncat ke iterasi berikutnya

```python
while True:
    break
    print("Hello")
```

```python
i = 0
while True:
    print("i awal", i)
    i += 1
    if i > 5:
        break
    #print("i akhir", i)
```

```python
n = 5
while n > 0:
    if n % 2 == 0:
        print(n)
        break
    n -= 1
print("loop ended.")
```

```python
#while dalam satu baris
while n > 0: n -= 1; print(n)
```
